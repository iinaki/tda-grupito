```python
import time
import matplotlib.pyplot as plt
import random

class Batalla:
    def __init__(self, tiempo, peso):
        self.tiempo = tiempo
        self.peso = peso

    def optimizar_batallas(lista_batallas):
        lista_batallas.sort(key=lambda x: (x.peso/x.tiempo) ,reverse=True)

        suma,tiempo = 0,0
        for b in lista_batallas:
            tiempo += b.tiempo
            suma += tiempo * b.peso

        return lista_batallas, suma

tamanos = []
tiempos = []
tiempos_lineal = []
tiempo_log = []

for x in range(10, 10000, 100):
  tamanos.append(x)

tamanos = [10, 100,500 ,700, 1000, 2000,4000, 5000,7000,8500, 10000]

def arr_random(n):
  batallas = []
  lista_tiempo = []
  for x in range(n):
    tiempo = random.randrange(1, 1000)
    batallas.append(Batalla( tiempo, random.randrange(1, 1000)))
    lista_tiempo.append(tiempo)

  return batallas, lista_tiempo


for tamano in tamanos:
    batallas, lista_tiempo = arr_random(tamano)

    #PRUEBA BATALLAS
    inicio = time.time()
    Batalla.optimizar_batallas(batallas)
    fin = time.time()
    tiempo_transcurrido = (fin - inicio) * 1000  # Convertir a milisegundos
    tiempos.append(tiempo_transcurrido)


    #O(n)
    inicio = time.time()
    suma = 0
    for b in batallas:
      suma += b.peso
    fin = time.time()
    tiempo_transcurrido = (fin - inicio) * 1000  # Convertir a milisegundos
    tiempos_lineal.append(tiempo_transcurrido)

plt.plot(tamanos, tiempos, marker='o', label = "optimizar batallas")
plt.plot(tamanos, tiempos_lineal, marker='o' , label = "O(n)")

plt.legend()

plt.xlabel('Cantidad de elementos')
plt.ylabel('Tiempo (ms)')
plt.title('Complejidad temporal')
plt.grid(True)
plt.show()

```
